```python
import random
import pandas as pd # A module for creating tables
import numpy as np # A module for fast numerical calculation

# Initializes all variables to store values
motor_random_number = []
gear_random_number = []
motor_unit_running_period = []
gear_unit_running_period = []
combined_unit_running_period = []
cum_running_period = []
no_of_failed_unit = []
cum_avg_running_period = []


for i in range(50):
    # Simulation for motor unit failure
    r = random.uniform(0,1)
    motor_random_number.append(round(r * 100))
    no_of_failed_unit.append(i + 1)
    
    if 0 < r <= 0.40:
        motor_unit_running_period.append(5)
    elif 0.40 < r <= 0.55:
        motor_unit_running_period.append(15)
    elif 0.55 < r <= 0.65:
        motor_unit_running_period.append(25)
    elif 0.65 < r <= 0.75:
        motor_unit_running_period.append(35)
    elif 0.75 < r <= 0.95:
        motor_unit_running_period.append(45)
    else:
        motor_unit_running_period.append(55)
        
    # Simulation for gear unit failure
    r = random.uniform(0,1)
    gear_random_number.append(round(r * 100))
    
    if 0 < r <= 0.07:
        gear_unit_running_period.append(5)
    elif 0.07 < r <= 0.27:
        gear_unit_running_period.append(15)
    elif 0.27 < r <= 0.60:
        gear_unit_running_period.append(25)
    elif 0.60 < r <= 0.87:
        gear_unit_running_period.append(35)
    else:
        gear_unit_running_period.append(45)
    

#Calculates the combined running unit period  
combined_unit_running_period = np.minimum(motor_unit_running_period, gear_unit_running_period)

#Calculates the cumulative running period
cum_running_period = np.cumsum(combined_unit_running_period)

#Calculates the average of cumulative running period by dividing with no of failed unit
temp_var = np.divide(cum_running_period, no_of_failed_unit)
cum_avg_running_period = np.round(temp_var, 1)

simulation_table = pd.DataFrame({"Motor Random Number": motor_random_number,
                                "Motor Unit Running Period": motor_unit_running_period,
                                "Gear Random Number": gear_random_number,
                                "Gear Unit Running Period": gear_unit_running_period,
                                "No of Failed Unit": no_of_failed_unit,
                                "Combined Unit Running Period": combined_unit_running_period,
                                "Cumulative Running Period": cum_running_period,
                                "Cumulative Average Running Period": cum_avg_running_period})
```

```python
# Calculates and displays the Frequemcy table
weeks_btw_failure = simulation_table['Combined Unit Running Period'].value_counts().index
frequency = simulation_table['Combined Unit Running Period'].value_counts().values

freq_table = pd.DataFrame({"Weeks between Failure": weeks_btw_failure,
                          "Frequency": frequency})

freq_table
```
