```python
program simulation_program_1
implicit none

!Initialize all variables to use
integer count,i
integer, dimension(1000) :: seed
real r
real, dimension(1000) :: no_of_failed_unit, motor_rand_numbers, motor_unit_running_period,cumm_running_period
real, dimension(1000) :: gear_random_number, gear_unit_running_period,cumm_avg_running_period,combined_unit_running_period

!Initialize variables to hold the count
integer, dimension(1) :: k1 = 0, k2 = 0, k3 = 0, k4 = 0, k5 = 0


!Set system clock so we can generate unique random number at every run
call system_clock(count)
seed = count
call random_seed(put = seed)


! Run the simulation for 1000 times
do i = 1,1000

    !Simulation for motor unit
    call random_number(r)
    no_of_failed_unit(i) = i + 1
    motor_rand_numbers(i) = int(r * 100)

    if (r <= 0.40) then
        motor_unit_running_period(i) = 5
    else if (r > 0.40 .and. r <= 0.55) then
        motor_unit_running_period(i) = 15
    else if (r > 0.55 .and. r <= 0.65) then
        motor_unit_running_period(i) = 25
    else if (r > 0.65 .and. r <= 0.75) then
        motor_unit_running_period(i) = 35
    else if (r > 0.75 .and. r <= 0.95) then
        motor_unit_running_period(i) = 45
    else
        motor_unit_running_period(i) = 55
    endif
    !Reset seed before start of second simulation
    call random_seed(get = seed)

    !Simulation for gear unit
    call random_number(r)
    gear_random_number(i) = int(r * 100)

    if (r <= 0.07) then
        gear_unit_running_period(i) = 5
    else if (r > 0.07 .and. r <= 0.27) then
        gear_unit_running_period(i) = 15
    else if (r > 0.27 .and. r <= 0.60) then
        gear_unit_running_period(i) = 25
    else if (r > 0.60 .and. r <= 0.87) then
        gear_unit_running_period(i) = 35
    else
        gear_unit_running_period(i) = 45
    endif

    ! Reset seed before next iteration
    call random_seed(get = seed)

    !Calculate the combined unit running time
    combined_unit_running_period(i) = MIN(motor_unit_running_period(i), gear_unit_running_period(i))

end do

!Calculate the cummulative average running period
cumm_avg_running_period(1) = combined_unit_running_period(1)
do i = 2,1000
    cumm_avg_running_period(i) = cumm_avg_running_period(i-1) + combined_unit_running_period(i)
end do

do i = 1,1000
    if (combined_unit_running_period(i) == 5) then
        k1 = k1 + 1
    else if (combined_unit_running_period(i) == 15)then
        k2 = k2 + 1
    else if (combined_unit_running_period(i) == 25)then
        k3 = k3 + 1
    else if (combined_unit_running_period(i) == 35)then
        k4 = k4 + 1
    else 
        k5 = k5 + 1
    endif
end do

!Print Fequency distribution table
print *,'Frequency of 5 weeks before failure: ', k1
print *,'Frequency of 15 weeks before failure: ', k2
print *,'Frequency of 25 weeks before failure: ', k3
print *,'Frequency of 35 weeks before failure: ', k4
print *,'Frequency of 45 weeks before failure: ', k5



print *,'---------------------------NO OF FAILED UNIT--------------------------------'
print *, no_of_failed_unit
print *,'------------------------------MOTOR RANDOM NUMBER--------------------------------'
print *, motor_rand_numbers
print *,'------------------------------GEAR RANDOM NUMBER--------------------------------'
print *, gear_random_number
print *,'-------------------------GEAR UNIT RUNNING TIME--------------------------------'
print *, gear_unit_running_period
print *,'------------------------MOTOR UNIT RUNNING TIME--------------------------------'
print *, motor_unit_running_period
print *,'------------------------COMBINED UNIT RUNNING TIME--------------------------------'
print *, combined_unit_running_period
print *,'---------------------CUMMULATIVE AVERAGE RUNNING PERIOD--------------------------------'
print *, cumm_avg_running_period


end program simulation_program_1
```
