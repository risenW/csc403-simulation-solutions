```python
package simulation_assignment;

public class simulation_problem_3 {

    public static void main(String[] args) {
        // Initialize variables
        int[] motor_random_number = new int[100];
        int[] gear_random_number = new int[100];

        int[] motor_unit_run_period = new int[100];
        int[] gear_unit_run_period = new int[100];

        int[] combined_unit_run_period = new int[100];
        int[] cumm_run_period = new int[100];

        int[] no_of_failed_unit = new int[100];
        double[] cum_avg_run_period = new double[100];

        int[] wks_btw_failure = {5, 15, 25, 35, 45, 55};
        int[] frequency = new int[6];

        double rand;

        //Simulation for motor unit
        for (int i = 0; i < 100; i++) {
            no_of_failed_unit[i] = i + 1;
            rand = Math.random();
            motor_random_number[i] = (int) (rand * 100);

            if (0 < rand && rand <= 0.40) {
                motor_unit_run_period[i] = 5;
            } else if (0.40 < rand && rand <= 0.55) {
                motor_unit_run_period[i] = 15;
            } else if (0.55 < rand && rand <= 0.65) {
                motor_unit_run_period[i] = 25;
            } else if (0.65 < rand && rand <= 0.75) {
                motor_unit_run_period[i] = 35;
            } else if (0.75 < rand && rand <= 0.94) {
                motor_unit_run_period[i] = 45;
            } else {
                motor_unit_run_period[i] = 55;
            }
            
            //Simulation for gear unit
            rand = Math.random();
            gear_random_number[i] = (int) (rand * 100);
            
            if (0 < rand && rand <= 0.07) {
                gear_unit_run_period[i] = 5;
            } else if (0.07 < rand && rand <= 0.27) {
                gear_unit_run_period[i] = 15;
            } else if (0.27 < rand && rand <= 0.60) {
                gear_unit_run_period[i] = 25;
            } else if (0.60 < rand && rand <= 0.87) {
                gear_unit_run_period[i] = 35;
            } else {
                gear_unit_run_period[i] = 45;
            }
        }

        //compute the combined unit running period
        combined_unit_run_period = get_min_array(motor_unit_run_period,gear_unit_run_period);

        //Compute the cumulative sum of the combined unit running period 
        cumm_run_period = get_cumm_sum(combined_unit_run_period);
        
        //compute the cumulative average running period 
        cum_avg_run_period = get_cumm_avg(cumm_run_period, no_of_failed_unit);
        
        //Getting the Frequency
        frequency = get_frequency(wks_btw_failure, combined_unit_run_period);

        
        System.out.println("Frequency of 5 weeks before failure: " + frequency[0] );
        System.out.println("Frequency of 15 weeks before failure: " + frequency[1] );
        System.out.println("Frequency of 25 weeks before failure: " + frequency[2] );
        System.out.println("Frequency of 35 weeks before failure: " + frequency[3] );
        System.out.println("Frequency of 45 weeks before failure: " + frequency[4] );


    }

    public static int[] get_min_array(int[] array1, int[] array2) {
        int[] temp_array = new int[array1.length];
        for (int i = 0; i < array1.length; i++) {
            temp_array[i] = ((array1[i] < array2[i]) ? array1[i] : array2[i]);
        }
        return temp_array;
    }

    public static int[] get_cumm_sum(int[] array1) {
        int[] temp_array = new int[array1.length];
        temp_array[0] = array1[0];
        for (int i = 1; i < array1.length; i++) {
            temp_array[i] = temp_array[i - 1] + array1[i];
        }
        return temp_array;
    }

    public static double[] get_cumm_avg(int[] array1, int[] array2) {
        double[] temp_array = new double[array1.length];
        for (int i = 0; i < array1.length; i++) {
            temp_array[i] = ((array1[i] / array2[i]) * 100) / 100.0;
        }
        return temp_array;
    }

    public static int[] get_frequency(int[] array1, int[] array2) {
        int[] temp_array = new int[6];
        for (int i = 0; i < temp_array.length; i++) {
            for (int j = 0; j < array2.length; j++) {
                if (array1[i] == array2[j]) {
                    temp_array[i] += 1;
                }
            }
        }
        return temp_array;
    }

}
```
