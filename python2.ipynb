```python
import random
import pandas as pd # A module for creating tables
import numpy as np # A module for fast numerical calculation

#Initialize all variables
day = []
random_numbers = []
polio_cases = []

# Simulation for polio cases
for i in range(50):
    r = random.uniform(0,1)
    random_numbers.append(round(r * 100))
    day.append(i + 1)
    
    if 0 < r <= 0.05:
        polio_cases.append(0)
    elif 0.05 < r <= 0.20:
        polio_cases.append(1)
    elif 0.20 < r <= 0.42:
        polio_cases.append(2)
    elif 0.42 < r <= 0.64:
        polio_cases.append(3)
    elif 0.64 < r <= 0.81:
        polio_cases.append(4)
    elif 0.81 < r <= 0.91:
        polio_cases.append(5)
    elif 0.91 < r <= 0.96:
        polio_cases.append(6)
    elif 0.96 < r <= 0.98:
        polio_cases.append(7)
    elif 0.98 < r <= 0.99:
        polio_cases.append(8)
    else:
        polio_cases.append(9)
    

simulation_table = pd.DataFrame({"Day": day,
                                "Random Number": random_numbers,
                                "Polio": polio_cases})
#Displays the table
simulation_table
```

```python
average = np.sum(simulation_table['Polio']) / len(polio_cases)
print("The average Polio outbreak is {}".format(average))
```
