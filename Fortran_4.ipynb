```python
program simulation_program_4
implicit none

integer count,i
real r, avg_waiting_time, avg_idle_time
real, dimension(1000) :: no_of_event, arrival_rand_numbers, inter_arrival_time
real, dimension(1000) :: maintenance_random_number, arrival
real, dimension(1000) :: service_time, service_start, service_end
real, dimension(1000) :: customer_wait_time, idle_time



integer, dimension(1000) :: seed

call system_clock(count)
seed = count
call random_seed(put = seed)


do i = 1,1000
    !Simulation for Inter Arrival time
    call random_number(r)
    no_of_event(i) = i + 1
    arrival_rand_numbers(i) = int(r * 100)

    if (r <= 0.50) then
        inter_arrival_time(i) = 1
    else if (r > 0.50 .and. r <= 0.80) then
        inter_arrival_time(i) = 2
    else
        inter_arrival_time(i) = 3
    endif
    !Reset seed
    call random_seed(get = seed)

    ! Simulation for Maintenance schedule
    call random_number(r)
    maintenance_random_number(i) = int(r * 100)

    if (r <= 0.65) then
        service_time(i) = 1
    else if (r > 0.65 .and. r <= 0.95) then
        service_time(i) = 2
    else
        service_time(i) = 3
    endif

    call random_seed(get = seed)
end do

! This section computes the arrival time, service time,
! service start time, service end time, customer waiting time and idle time

!Default starting values
arrival(1) = inter_arrival_time(1)
service_start(1) = arrival(1)
service_end(1) = service_start(1) + service_time(1)
customer_wait_time(1) = 0
idle_time(1) = service_start(1)

!Computation of later values
do i = 2,1000
    arrival(i) = arrival(i-1) + inter_arrival_time(i)
    service_start(i) = max(service_end(i - 1), arrival(i))
    service_end(i) = service_start(i) + service_time(i)

    if (service_end(i - 1) < arrival(i)) then
        customer_wait_time(i) = 0
        idle_time(i) = arrival(i) - service_end(i - 1)
    else
        customer_wait_time(i) = service_end(i -1) - arrival(i)
        idle_time(i) = 0
    endif

end do

! Print the Averages
avg_waiting_time = sum(customer_wait_time) / 1000
avg_idle_time = sum(idle_time) / 1000

print *,'The Average customer waiting time is: ',avg_waiting_time
print *,'The Average idle time is: ',avg_idle_time


print *,'------------------NO OF FAILED EVENT--------------------------------'
print *, no_of_event
print *,'-----------------ARRIVAL RANDOM NUMBER--------------------------------'
print *, arrival_rand_numbers
print *,'------------------INTER-ARRIVAL TIME--------------------------------'
print *, inter_arrival_time
print *,'---------------------ARRIVAL TIME--------------------------------'
print *, arrival
print *,'----------------MAINTENANCE RANDOM NUMBER--------------------------------'
print *, maintenance_random_number
print *,'----------------------SERVICE TIME--------------------------------'
print *, service_time
print *,'---------------------SERVICE START--------------------------------'
print *, service_start
print *,'----------------------SERVICE END--------------------------------'
print *, service_end
print *,'---------------------CUSTOMER WAITING--------------------------------'
print *, customer_wait_time
print *,'-----------------------IDLE TIME--------------------------------'
print *, idle_time


end program simulation_program_4
```
