```python
import random
import pandas as pd # A module for creating tables
import numpy as np # A module for fast numerical calculation

#Initialize variables to hold the count
k1 = 0; k2 = 0; k3 = 0; k4 = 0; k5 = 0

random_numbers = []
test = []
no_of_survival = []

#Simulation for Number of Survivals
for i in range(50):
    r = random.uniform(0,1)
    random_numbers.append(round(r * 100))
    test.append(i + 1)
    
    if 0 < r <= 0.1296:
        no_of_survival.append(0)
        k1 = k1 + 1
    elif 0.1296 < r <= 0.4752:
        no_of_survival.append(1)
        k2 = k2 + 1
    elif 0.4752 < r <= 0.8208:
        no_of_survival.append(2)
        k3 = k3 + 1
    elif 0.8208 < r <= 0.9744:
        no_of_survival.append(3)
        k4 = k4 + 1
    else:
        no_of_survival.append(4)
        k5 = k5 + 1
    
        
print("In the first interval, the number of survivals is: {}".format(k1))
print("In the second interval, the number of survivals is: {}".format(k2))
print("In the third interval, the number of survivals is: {}".format(k3))
print("In the fourth interval, the number of survivals is: {}".format(k4))
print("In the fifth interval, the number of survivals is: {}".format(k5))

#Create the simulation table
simulation_table = pd.DataFrame({"Test": test,
                                "Random Number": random_numbers,
                                "Number of Survivals": no_of_survival}) 

#Display the table
simulation_table
```

```python
#Calculates the average number of survivals
average_no_survivals = np.sum(simulation_table['Number of Survivals']) / len(no_of_survival)
print("The Average number of survivals is {}".format(average_no_survivals))
```
