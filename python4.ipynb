```python
import random
import pandas as pd # A module for creating tables
import numpy as np # A module for fast numerical calculation

# Initializes all variables to store values
no_of_event = []
arrival_random_number = []
maintenance_random_number = []
inter_arrival_time = []
arrival = []
service_time = []
service_start = []
service_end = []
customer_wait_time = []
idle_time = []


for i in range(50):
    # Simulation for failure arrivals
    r = random.uniform(0,1)
    arrival_random_number.append(round(r * 100))
    no_of_event.append(i + 1)
    
    if 0 < r <= 0.50:
        inter_arrival_time.append(1)
    elif 0.50 < r <= 0.8:
        inter_arrival_time.append(2)
    else:
        inter_arrival_time.append(3)
        
     
    # Simulation for maintenance
    r = random.uniform(0,1)
    maintenance_random_number.append(round(r * 100))
    
    if 0 < r <= 0.65:
        service_time.append(1)
    elif 0.65 < r <= 0.95:
        service_time.append(2)
    else:
        service_time.append(3)
              

#Calculates arrival time as the cumulative inter-arrival-time 
arrival = np.cumsum(inter_arrival_time)


# This next code block computes the service time,
# service start time, service end time, customer waiting time and idle time
service_start.append(arrival[0])
service_end.append(service_start[0] + service_time[0])
customer_wait_time.append(0)
idle_time.append(service_start[0])

for i in range(1,50):    
    service_start.append(max(service_end[i - 1], arrival[i]))
    service_end.append(service_start[i] + service_time[i])
    
    if service_end[i - 1] < arrival[i]:
        customer_wait_time.append(0)
        idle_time.append(arrival[i] - service_end[i - 1])
    else:
        customer_wait_time.append(service_end[i - 1] - arrival[i])
        idle_time.append(0)
        
        
#Create the simulation table        
simulation_table = pd.DataFrame({"No of Event": no_of_event,
                                "Random Number (Arrival)": arrival_random_number,
                                "Inter Arrival Time": inter_arrival_time,
                                "Arrival": arrival,
                                "Random Number (Maintenance)": maintenance_random_number,
                                "Service Time": service_time,
                                "Service Start": service_start,
                                "Service End": service_end,
                                "Customer Waiting": customer_wait_time,
                                "Idle Time": idle_time})

```

```python
#Display the simulation table
simulation_table
```

```python

#Print the Averages
avg_waiting_time = sum(customer_wait_time) / len(no_of_event)
avg_idle_time = sum(idle_time) / len(no_of_event)

print("The Average customer waiting time is: {}".format(avg_waiting_time))
print("The Average Idle time is: {}".format(avg_idle_time))
```
