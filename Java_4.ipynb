```python
package simulation_assignment;

public class simulation_problem_4 {

    public static void main(String[] args) {
        int[] cust_random_number = new int[100];
        int[] service_random_number = new int[100];

        int[] inter_arrival_time = new int[100];
        int[] arrival = new int[100];

        int[] service_time = new int[100];
        int[] service_start = new int[100];

        int[] no_of_event = new int[100];
        int[] service_end = new int[100];
        int[] customer_waiting_time = new int[100];
        int[] server_waiting_time = new int[100];

        double rand = 0;
        double avg_waiting_time, avg_idle_time;

        for (int i = 0; i < 100; i++) {
            //Simulation for customer
            rand = Math.random();
            no_of_event[i] = (i + 1);
            cust_random_number[i] = (int) (rand * 100);

            if (0 < rand && rand <= 0.50) {
                inter_arrival_time[i] = 1;
            } else if (0.50 < rand && rand < 0.80) {
                inter_arrival_time[i] = 2;
            } else {
                inter_arrival_time[i] = 3;
            }

            //Simulation for server
            rand = Math.random();
            service_random_number[i] = (int) (rand * 100);
            if (0 < rand && rand <= 0.65) {
                service_time[i] = 1;
            } else if (0.65 < rand && rand <= 0.95) {
                service_time[i] = 2;
            } else {
                service_time[i] = 3;
            }
        }

        /* ! This section computes the arrival time, service time,
          service start time, service end time, customer waiting time and idle time */
        // START
        // Enter default starting values to te arrays
        arrival[0] = inter_arrival_time[0];
        service_start[0] = arrival[0];
        service_end[0] = service_start[0] + service_time[0];
        customer_waiting_time[0] = 0;
        server_waiting_time[0] = service_time[0];

        for (int i = 1; i < service_start.length; i++) { //start the loop from 1
            arrival[i] = arrival[i - 1] + inter_arrival_time[i];
            service_start[i] = get_min_num(service_end[i - 1], arrival[i]);
            service_end[i] = service_start[i] + service_time[i];

            if (service_end[i - 1] < arrival[i]) {
                customer_waiting_time[i] = 0;
                server_waiting_time[i] = arrival[i] - service_end[i - 1];
            } else {
                server_waiting_time[i] = 0;
                customer_waiting_time[i] = service_end[i - 1] - arrival[i];
            }

        }
        // END

        //Print the averages
        avg_waiting_time = get_sum_of_array(customer_waiting_time) / no_of_event.length;
        avg_idle_time = get_sum_of_array(server_waiting_time) / no_of_event.length;

        System.out.println("The Average customer waiting time is: " + avg_waiting_time);
        System.out.println("The Average idle time is: " + avg_idle_time);

    }

    public static int[] get_cumm_sum(int[] array1) {
        /* Returns the cumulative sum of array*/
        int[] array3 = new int[array1.length];
        array3[0] = array1[0];
        for (int i = 1; i < array1.length; i++) {
            array3[i] = array3[i - 1] + array1[i];
        }
        return array3;
    }

    public static int get_min_num(int a, int b) {
        /* Returns minimum of two integer values*/
        if (a > b) {
            return a;
        } else {
            return b;
        }
    }

    public static int get_sum_of_array(int[] array) {
        //A method to sum all the elements of an array
        int sum = 0;
        for (int i = 0; i < array.length; i++) {
            sum += array[i];
        }
        return sum;
    }
}
```
